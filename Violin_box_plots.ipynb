```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format='retina'
```

```python


data = [90,60,75,48.57,50.00,26.47,50.,50.,7.14,2.08]

fig, ax = plt.subplots(figsize=(6,5))
ax.boxplot(data, labels=['Media regrowth'])

vio = ax.violinplot(data, widths=0.4)
ax.set_ylabel('percentage');
ax.set_ylim(0,100)
vio['cbars'].set_alpha(0)
vio['cmins'].set_color('black')
vio['cmins'].set_linewidth(1)
vio['cmaxes'].set_color('black')
vio['cmaxes'].set_linewidth(1)
plt.show()
```
